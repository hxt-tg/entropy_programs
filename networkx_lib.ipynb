## 构建网络示例代码如下

````python
# imports
import matplotlib.pyplot as plt
import networkx as nx
````

````python
G = nx.Graph()
G.add_nodes_from([1, 2, 3])
G.add_edges_from([(1, 2), (1, 3)])
print('#Nodes:', G.number_of_nodes(), '  #Edges:', G.number_of_edges())
print('Nodes:', list(G.nodes))
print('Edges:', list(G.edges))
print('Neighbors of node 1:', list(G.neighbors(1)))
print('Degree of node 1:', G.degree[1])
G.remove_node(2)
print('Nodes:', list(G.nodes))
print('Edges:', list(G.edges))
G.clear()
````

上面的代码中，`nx.Graph()`用于创建一个空网络，`G.add_node(1)`用于添加节点`1`，参数可以是数字或者字符串。
`G.add_nodes_from([1, 2])`则传入列表以添加节点。
添加边同理，使用`G.add_edge(1, 2)`或者`G.add_edges_from([(1, 2), (1, 3)])`，参数相应变为一对节点。

删除节点和边则将这一系列的函数中`add`替换为`remove`即可。

获取节点相关信息的方式

| 对象元素 | 语法元素 | 含义 |
| :-----: | :-----: | :--- |
|`G.number_of_nodes()`|方法|获取网络的总节点数。|
|`G.number_of_edges()`|方法|获取网络的总边数。|
|`G.nodes`|属性(列表)|获取网络的所有节点。|
|`G.edges`|属性(列表)|获取网络的所有边。|
|`G.neighbors(i)`|方法|获取节点`i`的所有邻居节点标签或编号。|
|`G.adj`|属性(字典)|获取网络的邻接矩阵。|
|`G.degree`|属性(列表)|获取网络的节点度信息。|

注：`G.adj[i]`与`G.neighbors(i)`等效，但`G.adj`携带边信息，例如边权重等。

## 网络可视化如下


````python
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1, 4), (2, 4), (3, 4), (1, 2), (2, 3)])
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
````

## 构建星状网络
`nx.star_graph(n)`

构建从0到n(包含编号n)的n+1个节点的星状网络，第一个节点(0号节点)为中心节点。

````python
G = nx.star_graph(10)
nx.draw(G, with_labels=True, font_weight='bold')
plt.show()
````

## 构建规则网络
`G = nx.watts_strogatz_graph(n, d, p)`

WS小世界网络是基于规则网络随机重连边生成的，该方法生成一个由$n$个节点，每个节点的度都是$d$，且断边重连概率为$p$的WS小世界网络。

当断边重连不发生，即$p=0$时，该网络为规则网络。

节点编号从0至$n-1$。

````python
G = nx.watts_strogatz_graph(10, 4, 0.3)
nx.draw_circular(G, with_labels=False, node_size=30)
plt.show()
````

## 构建BA无标度网络
`G = nx.barabasi_albert_graph(n, m)`

构建一个初始连边为0的BA无标度网络，其终止状态总结点数为$n$，每次增长连边为$m$，$n>m$。

节点编号从0至$n-1$。

````python
G = nx.barabasi_albert_graph(50, 2)
nx.draw(G, with_labels=False, font_weight='bold', node_size=30)
plt.show()
````

## 构建穴居人网络
`G = nx.caveman_graph(m, n)`

构建m个n节点全连通的穴居人网络。

````python
G = nx.caveman_graph(3, 5)
nx.draw(G, with_labels=False, font_weight='bold', node_size=1)
plt.show()
````
